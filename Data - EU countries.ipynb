So, in an extremely simple model, we have an equation in a form:
coeff[0] * gdp[2010] = migrants[2010]
coeff[0] * gdp[2015] = migrants[2015]
coeff[0] * gdp[2018] = migrants[2019]

and then we set coeff[0] to minimize squared errors
PLEASE NOTE THAT WE DON'T HAVE gdp data for 2019 year (only 2018)!

when we add another factor, e.g. underemployment it would look like this:
coeff[0] * gdp[2010] + coeff[1] * uempl[2010] = migrants[2010]
coeff[0] * gdp[2015] + coeff[1] * uempl[2015] = migrants[2015]
coeff[0] * gdp[2018] + coeff[1] * uempl[2018] = migrants[2019]
etc.
